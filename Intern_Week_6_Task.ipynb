# **Loading and Splitting the document**

```python
from langchain.document_loaders import DirectoryLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
import os

def load_docs(directory):
    """
    Load documents from the specified directory.

    Args:
        directory (str): Path to the directory containing documents.

    Returns:
        list: List of loaded documents.
    """
    if not os.path.isdir(directory):
        print(f"Error: Directory '{directory}' does not exist.")
        return []

    print(f"Loading documents from '{directory}'...")
    loader = DirectoryLoader(directory)
    documents = loader.load()

    if not documents:
        print("Warning: No documents found in the directory.")

    return documents

```

```python
def split_docs(documents, chunk_size=1000, chunk_overlap=200):
    """
    Split documents into chunks of text.

    Args:
        documents (list): List of documents to be split.
        chunk_size (int, optional): Maximum size of each chunk of text. Defaults to 1000.
        chunk_overlap (int, optional): Number of characters by which adjacent chunks overlap. Defaults to 200.

    Returns:
        list: List of split documents.
    """
    # Initialize RecursiveCharacterTextSplitter
    text_splitter = RecursiveCharacterTextSplitter(chunk_size=chunk_size, chunk_overlap=chunk_overlap)

    try:
        # Split documents into chunks
        split_docs = text_splitter.split_documents(documents)
        print(f"Number of split documents: {len(split_docs)}")
        if split_docs:
            print(f"Length of content in the first split document: {len(split_docs[0].page_content)}")
        return split_docs
    except Exception as e:
        print(f"Error occurred during document splitting: {str(e)}")
        return []
```

```python
def main_load_and_split_docs(directory):
    documents = load_docs(directory)

    print(f"Number of documents loaded: {len(documents)}")

    if documents:
        print(f"Content of the first document:\n{documents[0].page_content[:1000]}")
        return split_docs(documents)
    else:
        print("Warning: No documents loaded.")

```

# **Importing Chroma for Vector DB**

```python
from langchain_chroma import Chroma
from langchain_community.embeddings import HuggingFaceHubEmbeddings

vectorstore = Chroma.from_documents(documents=split_documents, embedding=HuggingFaceHubEmbeddings())
```

# **Building a Retriever**

```python
retriever = vectorstore.as_retriever(search_type="similarity", search_kwargs={"k": 3})
```

```python
import textwrap

# Define a function to print the answer with formatted lines
def print_answer(answer):
    # Split the answer into lines with a maximum width of 60 characters
    lines = textwrap.wrap(answer, width=100)

    # Print each line
    for line in lines:
        print(line)

# Assuming you have a retriever system set up and imported as vectorstore
# Assuming you have set up a function to retrieve documents given a question

def retrieve_documents(question, k=5):
    # Use the retriever to retrieve documents related to the question
    documents = vectorstore.similarity_search(question, k=k)
    return documents

# Define questions
questions = [
    "Who was Bhaskar Varma?",
    "What social reforms did Jayasthiti Malla introduce?",
    "what was effects of Hari Singh Deva's conquest of Nepal?",
    "Who is Prithivi Narayan Shah?",
    "Who were Yaksha Malla's sond and daugthers?",
    "What is the legend of Manjushree and how does it relate to the geographical origins of Nepal?",
    "Why is the Lichchhavi period considered a golden age in Nepal’s history?",
    "When did ashoka visited Nepal?",
    "When did Prithivi Narayan Shah conquer kantipur?",
    "what were the names of dynasties in nepal?"
]

# Define a function to retrieve documents for a question and print the question along with the top document content
def retrieve_and_print_documents(question, k=1):
    # Retrieve documents for the question
    documents = retrieve_documents(question, k)

    # Print the question
    print("Question:", question)

    # Iterate through retrieved documents
    for i, document in enumerate(documents):
        # Get the content of the document
        document_content = document.page_content

        # Print the top document content
        print("Answer:")
        print_answer(document_content)
        print("\n")

# Iterate through questions
for question in questions:
    # Retrieve and print documents for the question
    retrieve_and_print_documents(question)
```

# **Build The Model Pipeline**

```python
from langchain.llms import HuggingFacePipeline

def create_language_model_pipeline(model_id, task, model_kwargs=None, device=-1):
    """
    Create a pipeline for interacting with a pre-trained language model from Hugging Face.

    Args:
        model_id (str): Identifier or name of the pre-trained language model to use.
        task (str): Task for which the pipeline will be used.
        model_kwargs (dict, optional): Additional keyword arguments to customize the behavior of the model.
        device (int, optional): Device on which the model will be run. -1 for CPU, 0 or more for GPU.

    Returns:
        HuggingFacePipeline: Pipeline object for interacting with the model.
    """
    try:
        pipeline = HuggingFacePipeline.from_model_id(
            model_id=model_id,
            task=task,
            model_kwargs=model_kwargs,
            device=device
        )
        return pipeline
    except Exception as e:
        print(f"Error creating language model pipeline: {str(e)}")
        return None
```

```python
from langchain.memory import ConversationBufferMemory
from langchain.chains import ConversationalRetrievalChain
import textwrap

def print_answer(answer):
    # Split the answer into lines with a maximum width of 60 characters
    lines = textwrap.wrap(answer, width=60)

    # Print each line
    for line in lines:
        print(line)

def main():
    model_id = "google/flan-t5-large"
    task = "text2text-generation"
    model_kwargs = {"max_length": 1058, "do_sample": False}  # Adjust max_length here
    device = -1  # -1 for CPU, 0 or more for GPU

    language_model_pipeline = create_language_model_pipeline(model_id, task, model_kwargs, device)
    if language_model_pipeline:
        print("Language model pipeline created successfully.")
        # Use the pipeline for further interactions

        # Create memory
        memory = ConversationBufferMemory(llm=language_model_pipeline, memory_key="chat_history", output_key='answer', return_messages=True)

        # Create and use the ConversationalRetrievalChain
        chain = ConversationalRetrievalChain.from_llm(language_model_pipeline, retriever=retriever, memory=memory)

        while True:
            user_input = input("User: ")
            if user_input.lower() == "exit":
                print("Chatbot: Thanks!")
                break
            result = chain.invoke({"question": user_input})  # Use invoke instead of __call__

            # Access the correct key for the answer
            answer = result["answer"]

            print("Chatbot:")
            print_answer(answer)  # Print the answer with line breaks

    else:
        print("Failed to create language model pipeline.")

if __name__ == "__main__":
    main()
```
